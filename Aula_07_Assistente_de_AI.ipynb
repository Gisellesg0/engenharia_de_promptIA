### Calculadora Simples em Python

Este código demonstra uma calculadora básica que realiza operações de adição, subtração, multiplicação e divisão usando instruções `if` e `else`.

```python
def calculadora():
    print("Selecione a operação:")
    print("1. Adição")
    print("2. Subtração")
    print("3. Multiplicação")
    print("4. Divisão")

    while True:
        escolha = input("Digite sua escolha (1/2/3/4): ")

        # Verifica se a escolha da operação é válida
        if escolha in ('1', '2', '3', '4'):
            try:
                # Pede os números ao usuário e tenta convertê-los para float
                num1 = float(input("Digite o primeiro número: "))
                num2 = float(input("Digite o segundo número: "))
            except ValueError:
                # Lida com entrada inválida (não numérica)
                print("Entrada inválida. Por favor, digite um número.")
                continue # Continua para a próxima iteração do loop

            # Executa a adição
            if escolha == '1':
                print(f"{num1} + {num2} = {num1 + num2}")

            # Executa a subtração
            elif escolha == '2':
                print(f"{num1} - {num2} = {num1 - num2}")

            # Executa a multiplicação
            elif escolha == '3':
                print(f"{num1} * {num2} = {num1 * num2}")

            # Executa a divisão
            elif escolha == '4':
                # Verifica se o divisor é zero para evitar erro
                if num2 == 0:
                    print("Erro! Divisão por zero.")
                else:
                    print(f"{num1} / {num2} = {num1 / num2}")

            # Pergunta se o usuário deseja realizar outro cálculo
            proximo_calculo = input("Deseja fazer outro cálculo? (sim/não): ")
            if proximo_calculo.lower() == "não":
                break # Sai do loop se o usuário não quiser continuar
        else:
            # Mensagem de erro para escolha de operação inválida
            print("Entrada inválida. Por favor, digite 1, 2, 3 ou 4.")
```

### Demonstração da Calculadora

```python
# Para rodar a calculadora, chame a função:
calculadora()
```

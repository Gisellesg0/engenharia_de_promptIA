Para verificar a temperatura do sistema, precisaremos de uma biblioteca que possa acessar essas informações. A biblioteca `psutil` é uma ótima opção para isso, pois fornece uma interface para informações de processos e utilização do sistema (CPU, memória, discos, rede, sensores).

```
# Instala a biblioteca psutil, caso ainda não esteja instalada.
!pip install psutil
```

```
import psutil # Importa a biblioteca psutil para acesso a informações do sistema

def check_temperature(threshold=30): # Define a função check_temperature com um limite padrão de 30 graus Celsius
    """
    Verifica a temperatura do sistema e exibe um alerta se estiver acima do limite.
    Args:
        threshold (int): O limite de temperatura em graus Celsius.
    """
    temperatures = psutil.sensors_temperatures() # Obtém todas as leituras de sensores de temperatura do sistema

    # Verifica se há leituras de temperatura disponíveis
    if not temperatures: # Se a lista de temperaturas estiver vazia
        print("Não foi possível obter leituras de temperatura do sistema.") # Imprime uma mensagem de erro
        return # Sai da função

    # Tenta encontrar uma temperatura principal (como CPU)
    cpu_temp = None # Inicializa a variável para a temperatura da CPU como None
    for name, entries in temperatures.items(): # Itera sobre os diferentes tipos de sensores (ex: 'coretemp')
        for entry in entries: # Itera sobre as entradas de temperatura para cada sensor
            # Verifica se o rótulo da entrada contém 'cpu' ou 'core' (ignorando maiúsculas/minúsculas)
            if 'cpu' in entry.label.lower() or 'core' in entry.label.lower():
                cpu_temp = entry.current # Atribui a temperatura atual à cpu_temp
                break # Sai do loop interno, pois encontramos uma temperatura da CPU
        if cpu_temp is not None: # Se a temperatura da CPU foi encontrada
            break # Sai do loop externo

    if cpu_temp is None and temperatures: # Se nenhuma temperatura de CPU específica foi encontrada, mas há outras temperaturas
        # Se não encontrou uma temperatura de CPU específica, pega a primeira disponível
        first_entry = list(temperatures.values())[0][0] # Pega a primeira entrada de temperatura disponível
        cpu_temp = first_entry.current # Atribui a temperatura atual dessa entrada à cpu_temp
        print("Usando a temperatura de '" + str(first_entry.label) + "' como referência.") # Informa qual temperatura está sendo usada

    if cpu_temp is not None: # Se uma temperatura válida foi extraída
        # Imprime a temperatura atual formatada para duas casas decimais
        print("Temperatura atual do sistema: " + "{:.2f}".format(cpu_temp) + "°C")
        if cpu_temp > threshold: # Se a temperatura da CPU for maior que o limite
            # Imprime um alerta de superaquecimento com a temperatura e o limite
            print("ALERTA: Superaquecimento detectado! Temperatura (" + "{:.2f}".format(cpu_temp) + "°C) acima do limite de " + str(threshold) + "°C.")
        else: # Se a temperatura da CPU não for maior que o limite
            print("Temperatura do sistema está normal.") # Imprime que a temperatura está normal
    else: # Se não foi possível extrair nenhuma temperatura válida
        print("Não foi possível extrair uma temperatura válida do sistema.") # Imprime uma mensagem de erro
```

```
# Chama a função para verificar a temperatura
check_temperature(threshold=30)
```
